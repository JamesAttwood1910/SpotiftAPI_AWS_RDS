```python
# SQL queries
```

```python
# 1) Select 10 artists and their genres
```

```python
sql = "select artist, genre from spotify2 LIMIT 10"
cursor.execute(sql)
cursor.fetchall()
```

```python
# 2) Select all information for the 5 most popular J Balvin songs
```

```python
sql = "select * from spotify2 where artist = 'J Balvin' ORDER BY popularity DESC LIMIT 5"
cursor.execute(sql)
cursor.fetchall()
```

```python
# 3) Select all information for 5 most popular songs with popularity greater than 50 but less than 60. 
```

```python
sql = "select * from spotify2 where popularity > 50 AND popularity < 60 ORDER BY popularity DESC LIMIT 5"
cursor.execute(sql)
cursor.fetchall()
```

```python
# 4) select all information for songs with release date before 2020/02/22
```

```python
sql = "select * from spotify2 where release_date < '2020-02-22 00:00:00.000'"
cursor.execute(sql)
cursor.fetchall()
```

```python
# 5) count number of songs by genre.
```

```python
sql = "SELECT genre, COUNT(popularity) FROM spotify2 GROUP BY genre"
cursor.execute(sql)
cursor.fetchall()
```

```python
# Select 5 artists with most popular songs in total. 
```

```python
sql = "SELECT artist, genre, SUM(popularity) FROM spotify2 GROUP BY artist ORDER BY SUM(popularity) DESC LIMIT 5"
cursor.execute(sql)
cursor.fetchall()
```

```python
# select 5 artists with highest average song popularity with artist having over 4 songs in table. 
```

```python
sql = "SELECT artist, AVG(popularity), COUNT(artist) FROM spotify2 GROUP BY artist HAVING COUNT(artist) > 4 ORDER BY AVG(popularity) DESC LIMIT 5"
cursor.execute(sql)
cursor.fetchall()
```
